### Using Numpy and Pandas with Matplotlib

```python
import numpy as np
import pandas as pd
import random as rd
```

### Creating a array with random values

```python
y = np.array([])
for i in range(100):
   y = np.append(y,rd.randint(0,i))
```

```python
df = pd.DataFrame({"Numbers": y})
```

```python
fig, ax = plt.subplots()
ax.bar(df.index,df["Numbers"])
ax.set_title("Bar graph")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.show()
```
